```python
# dropping null values of product and Brand
df.dropna(subset=['product', 'brand'], inplace=True)

```

```python
# Replace null values with mean
df['sale_price'].fillna(df['sale_price'].mean(), inplace=True)
df['rating'].fillna(df['rating'].mean(), inplace=True)
```

```python
# Calculating the IQR for the "sale_price"
Q1_sale = df['sale_price'].quantile(0.25)
Q3_sale = df['sale_price'].quantile(0.75)
IQR_sale = Q3_sale - Q1_sale

lower_bound_sale = Q1_sale - 1.5 * IQR_sale
upper_bound_sale = Q3_sale + 1.5 * IQR_sale

# Replacing the outliers with the mean
mean_price = df['sale_price'].mean()
df['sale_price'] = np.where(
    (df['sale_price'] < lower_bound_sale) | (df['sale_price'] > upper_bound_sale),
    mean_price,
    df['sale_price']
)
```

```python
sns.boxplot(y=df['market_price'])
plt.title('Box plot for market price')
```

```python
Q1_market = df['market_price'].quantile(0.25)
Q3_market = df['market_price'].quantile(0.75)
IQR_market = Q3_market - Q1_market

lower_bound_market = Q1_market - 1.5 * IQR_market
upper_bound_market = Q3_market + 1.5 * IQR_market

mean_market = df['market_price'].mean()
df['market_price'] = np.where(
    (df['market_price'] < lower_bound_market) | (df['market_price'] > upper_bound_market),
    mean_market,
    df['market_price']
)
```

```python
sns.boxplot(y=df['market_price'])
plt.title('Box plot for market price')
```

```python
top_discount = df.groupby('product')['discount_percent'].max().sort_values(ascending=False).head(5)

# Plot Lollipop Chart
plt.figure(figsize=(10, 5))
plt.hlines(y=top_discount.index, xmin=0, xmax=top_discount.values, color='skyblue')
plt.plot(top_discount.values, top_discount.index, "o", color='blue')

plt.title('Top 5 Products by Average Discount')
plt.xlabel('Average Discount percent')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

```python
# Step 1: Filter products with discount_percent > 0
filtered_df = df[df['discount_percent'] > 0]
# Step 2: Group by product and get the max discount percent
least_discount = (filtered_df.groupby('product')['discount_percent'].max().sort_values().head(5))
# Step 3: Plot Lollipop Chart
plt.figure(figsize=(10, 5))
plt.hlines(y=least_discount.index, xmin=0, xmax=least_discount.values, color='skyblue')
plt.plot(least_discount.values, least_discount.index, "o", color='blue')
plt.title('Bottom 5 Products by Discount Percent')
plt.xlabel('Discount Percent')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```
